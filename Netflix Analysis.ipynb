```python
fig, axs = plt.subplots(2, 3, figsize=(18, 10))

type_counts = df["type"].value_counts()
axs[0,0].bar(type_counts.index, type_counts.values, color="#1f77b4")
axs[0,0].set_title("Movies vs TV Shows", fontsize=13, fontweight="bold")

country_counts = df["country"].value_counts().head(10)
axs[0,1].bar(country_counts.index, country_counts.values, color="#1f77b4")
axs[0,1].set_title("Top 10 Countries", fontsize=13, fontweight="bold")
axs[0,1].tick_params(axis="x", rotation=45)

axs[0,2].pie(type_counts.values, labels=type_counts.index, autopct="%1.1f%%",
             colors=["#1f77b4","#ff7f0e","#2ca02c"])
axs[0,2].set_title("Content Type Split", fontsize=13, fontweight="bold")

rating_counts = df["rating"].value_counts().head(8)
axs[1,0].bar(rating_counts.index, rating_counts.values, color="#1f77b4")
axs[1,0].set_title("Rating Distribution", fontsize=13, fontweight="bold")
axs[1,0].tick_params(axis="x", rotation=45)

release_counts = df["release_year"].value_counts().head(10)
axs[1,1].bar(release_counts.index, release_counts.values, color="#1f77b4")
axs[1,1].set_title("Top Release Years", fontsize=13, fontweight="bold")
axs[1,1].tick_params(axis="x", rotation=45)

genre_counts = df["genre"].value_counts().head(8)
axs[1,2].barh(genre_counts.index, genre_counts.values, color="#1f77b4")
axs[1,2].set_title("Top Genres", fontsize=13, fontweight="bold")

fig.suptitle("Netflix Dashboard 1 — Content Overview", fontsize=22, fontweight="bold")
plt.tight_layout()
plt.savefig("netflix_dashboard1.png", bbox_inches="tight", dpi=150)  # ← BEFORE show
plt.show()
print("✅ Dashboard 1 saved!")
```

```python
fig, axs = plt.subplots(2, 3, figsize=(18, 10))

movies = df[df["type"] == "Movie"]
tvshows = df[df["type"] == "TV Show"]

year_trend = df.groupby("release_year").size()
axs[0,0].plot(year_trend.index, year_trend.values, color="#1f77b4", linewidth=2.5, marker="o", markersize=4)
axs[0,0].set_title("Yearly Content Trend", fontsize=13, fontweight="bold")

movie_trend = movies["release_year"].value_counts().sort_index()
axs[0,1].plot(movie_trend.index, movie_trend.values, color="#1f77b4", linewidth=2.5, marker="o", markersize=4)
axs[0,1].set_title("Movie Release Trend", fontsize=13, fontweight="bold")

tv_trend = tvshows["release_year"].value_counts().sort_index()
axs[0,2].plot(tv_trend.index, tv_trend.values, color="#1f77b4", linewidth=2.5, marker="o", markersize=4)
axs[0,2].set_title("TV Show Release Trend", fontsize=13, fontweight="bold")

axs[1,0].hist(df["release_year"], bins=20, color="#1f77b4", edgecolor="white")
axs[1,0].set_title("Release Year Distribution", fontsize=13, fontweight="bold")

movie_genre = movies["genre"].value_counts().head(6)
axs[1,1].bar(movie_genre.index, movie_genre.values, color="#1f77b4")
axs[1,1].set_title("Top Movie Genres", fontsize=13, fontweight="bold")
axs[1,1].tick_params(axis="x", rotation=45)

tv_genre = tvshows["genre"].value_counts().head(6)
axs[1,2].bar(tv_genre.index, tv_genre.values, color="#1f77b4")
axs[1,2].set_title("Top TV Show Genres", fontsize=13, fontweight="bold")
axs[1,2].tick_params(axis="x", rotation=45)

fig.suptitle("Netflix Dashboard 2 — Trend Analysis", fontsize=22, fontweight="bold")
plt.tight_layout()
plt.savefig("netflix_dashboard2.png", bbox_inches="tight", dpi=150)  # ← BEFORE show
plt.show()
print("✅ Dashboard 2 saved!")
```
